```python
import requests
import re
import time
import os
header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36",
        "Connection": "keep-alive",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "zh-CN,zh;q=0.8"}
```

```python
def download_imgs(keyword):
    base_url1 = 'https://image.baidu.com/search/flip?tn=baiduimage&ie=utf-8&word={}&pn='.format(keyword)
    base_url2 = '&gsm=&ct=&ic=0&lm=-1&width=0&height=0'

    count = 0

    for i in range(5):
        page_num = str(i*20*3)
        url = base_url1 + page_num + base_url2
        response = requests.get(url, headers = header)
        picture_list = re.findall('{"thumbURL":"(.*?)",', response.text)
        for image_link in picture_list:
            print(image_link)
            count += 1
            connected = False
            while not connected == True:
                time.sleep(2)
                try:
                    img = requests.get(image_link, headers = header).content
                    connected = True
                except:
                    connected = False
            try:
                with open( './img/t/{}{}.jpg'.format(keyword,str(count)),'wb' ) as f: # path
                    f.write(img)
            except:
                os.mkdir('./img/t')  # path
                with open( './img/t/{}{}.jpg'.format(keyword,str(count)),'wb' ) as f: #path
                    f.write(img)
                
    return count
```

```python
keyword_list = ['','']
for name in keyword_list:
    print(download_imgs(name))
```
